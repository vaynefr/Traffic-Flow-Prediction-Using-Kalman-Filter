```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
data= pd.read_csv('Traffic.csv')
data.head()
```

```python
#dropping null rows
data=data.dropna()
```

```python
# Line plot of total traffic volume over time
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Total'])
plt.title('Total Traffic Volume Over Time')
plt.xlabel('Time')
plt.ylabel('Total Traffic Volume')
plt.show()
```

```python
# Bar plot of traffic situation
plt.figure(figsize=(12, 6))
data['Traffic Situation'].value_counts().plot(kind='bar')
plt.title('Traffic Situation')
plt.xlabel('Situation')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot of traffic situation
plt.figure(figsize=(12, 6))
data['Date'].value_counts().plot(kind='bar')
plt.title('Date')
plt.xlabel('Situation')
plt.ylabel('Count')
plt.show()
```

```python
# Line plot of total traffic volume over time
plt.figure(figsize=(12, 6))
plt.plot(data['Total'], data['Traffic Situation'])
plt.title('Total Traffic Volume Over Time')
plt.xlabel('Total Traffic Flow')
plt.ylabel('Traffic Situation')
plt.show()
```

```python
data['Traffic Situation'] = data['Traffic Situation'].map({'low': 0, 'normal': 1, 'heavy': 2})

# Calculate correlations
corr = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total', 'Traffic Situation']].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
data.hist(figsize=(15,12),bins = 15)
plt.title("Features Distribution")
plt.show()
```
